# Specify version of data to use and threshol RH value to signify clouds

```python

# VERSION 1:
# this was the original version, where RH for liquid only was used 
# to detect clouds, and then phase was assumed from temperature
# very low RH values (e.g. 90%) need to be used to detect reasonable cloud numbers
# corresponds to data ending in '_v1.nc'
#=============================================================
VERSION = 1
rh_thresh_liq = 90


# VERSION 2:
# I created this option after realizing air might be at saturation
# for ice but not liquid, so adapted to produce two RH values
# one relative to ice and one to liquid (this one turns out same as before)
# this shows clouds at ice relative humidities of 100%
# but probably too many clouds to be realistic
# also may be some actual bug that is messing up data @ different heights
# because some of the plots are patchy for some reason for winter data specifically
# corresponds to data ending in '_v2.nc'
#=============================================================
# VERSION = 2
# rh_thresh_liq = 90
# rh_thresh_ice = 100
```

### Plot relative humidity distributions (density histograms) as a function of height for each season from 12Z soundings 1997 - 2016

```python
# seasons to plot
# relative humidity threshold for clouds
#==================================================
seasons = ['autumn', 'winter', 'spring', 'summer']
#==================================================

# bin width, min edge, and max edge
# for RH distribution bins
#=====================================
width = 2
bin_min = 0
bin_max = 106
#=====================================
bin_edges = np.arange(bin_min,bin_max+width,width)
bin_centers = np.arange(bin_min+width/2,bin_max,width)


fig, axs = plt.subplots(1,len(seasons), figsize=(6.5,2))



for ss, season in enumerate(seasons):

    # open seasonal data
    ds = xr.open_dataset(f'./data/{season}_1997-2016_v{VERSION}.nc')

    # create density histogram at each height (integral across RH bins == 1)
    # across all non-nan values
    H = ds.h.values
    HIST = np.array([])
    
    # VERSION 1
    #----------
    if VERSION == 1:
        
        for hh, HH in enumerate(H):
            RH_vals = ds.RH.sel(h=HH).values
            the_hist = np.histogram(RH_vals[np.isnan(RH_vals) == False], bins = bin_edges, density=True);
            HIST = np.append(HIST, the_hist[0])
        axs[ss].vlines(rh_thresh_liq, 0, 6, colors='lightgray', linestyles='dashed', linewidths=0.5)
    
    # VERSION 2
    #----------
    elif VERSION == 2:
        
        # specify ice or liquid!
        #=======================
        VAR = 'ice'
#         VAR = 'liq'
        #=======================
        
        for hh, HH in enumerate(H):
            if str(VAR)=='ice':
                RH_vals = ds.RH_ice.sel(h=HH).values
            else:
                RH_vals = ds.RH_liq.sel(h=HH).values
            the_hist = np.histogram(RH_vals[np.isnan(RH_vals) == False], bins = bin_edges, density=True);
            HIST = np.append(HIST, the_hist[0])
        if str(VAR)=='ice':
            axs[ss].vlines(rh_thresh_liq, 0, 6, colors='lightgray', linestyles='dashed', linewidths=0.5)
        else:
            axs[ss].vlines(rh_thresh_ice, 0, 6, colors='lightgray', linestyles='dashed', linewidths=0.5)
            
        if ss == 0:
            print(f'plotting for {VAR}')
        
    # reshap histograms into grids
    HIST = np.reshape(HIST, (len(H),len(bin_centers)))
    rh, height = np.meshgrid(bin_centers, H)
    
    # plot distribution
    dens = axs[ss].pcolormesh(rh, height, HIST*100, vmin=0, vmax=6)
    
    # plot labels
    axs[ss].set_title(f'{season}')
    if ss == 0:
        axs[ss].set_ylabel('height (km)')
        axs[ss].set_xlabel('RH (%)')
    else:
        axs[ss].set_yticks([])
        axs[ss].set_xticks([50,100])
    
# colorbar
cbar_ax = fig.add_axes([0.925, 0.15, 0.025, 0.7])
fig.colorbar(dens, label='density ($\\times 10^{-2}$)', cax=cbar_ax)


# save
savename = f'./figures/RH_seasonalrange_{ds.year[0].values}-{ds.year[-1].values}.png'
print(f' >>> save as {savename}')
fig.savefig(savename, transparent=True, bbox_inches='tight', pad_inches=0, dpi=300)

```

### Plot frequency of RH threshold passed seasonall as a function of height

Various sources suggest temperature ranges where we might expect liquid vs supercooled water vs mixed ice/supercooled vs ice only only to exist in clouds:

- liquid: 0 C < T

- supercooled liquid: -15 C < T < 0 C

- mixed ice/supercooled: -40 C < T < -15 C

- ice: T < -40 C


Sources:

- https://www.skybrary.aero/articles/supercooled-water-droplets#:~:text=At%20temperatures%20between%200%C2%B0,crystals%20and%20supercooled%20water%20droplets.
- https://earthobservatory.nasa.gov/images/84916/supercool-clouds
- https://www.weather.gov/source/zhu/ZHU_Training_Page/clouds/cloud_development/clouds.htm

```python
#==================================================
seasons = ['autumn', 'winter', 'spring', 'summer']
colors = ['blue', 'dodgerblue', 'indianred', 'darkorange']
hours = [12]
show_cloud_type = True

# temperature ranges for cloud types
liq_range = (  0,  20)
sco_range = ( -5,  0 )
mix_range = (-25, -5 )
ice_range = (-60, -25)

#============================================================

fig, axs = plt.subplots(1,len(seasons), figsize=(10,2))

width = 2.5
bin_min = -45
bin_max = 20
bin_edges = np.arange(bin_min,bin_max+width,width)
bin_centers = np.arange(bin_min+width/2,bin_max,width)

for ss, season in enumerate(seasons):

    # open seasonal data
    ds = xr.open_dataset(f'./data/{season}_1997-2016_v{VERSION}.nc')

    # plot cloud occurence counts per total observations (times 100 to give % of total obs)
    
    # VERSION 1
    #----------
    if VERSION == 1:
        # find total counts of 'clouds' in different temperature bins
        hist_out = np.histogram(ds.T.values[ds.RH >= rh_thresh_liq], bins=bin_edges)
        #  find total obs 
        total_obs = (np.isnan(ds.RH)==False).sum(dim='doy').sum(dim='year').sum(dim='h').data
    
    # VERSION 2
    #----------
    elif VERSION == 2:
        # find total counts of 'clouds' in different temperature bins
        # when either cloud satisfying ice vapor RH or liq RH is found
        hist_out = np.histogram(ds.T.values[(ds.RH_ice >= rh_thresh_ice)+(ds.RH_liq >= rh_thresh_liq)], bins=bin_edges)
        #  find total obs # from liquid RH since ice RH set to zero when T > 0
        total_obs = (np.isnan(ds.RH_liq)==False).sum(dim='doy').sum(dim='year').sum(dim='h').data
    
    
    # plot bins
    if show_cloud_type == True:
        axs[ss].bar(bin_centers[(bin_centers<ice_range[1])&(bin_centers>ice_range[0])], 100*hist_out[0][(bin_centers<ice_range[1])&(bin_centers>ice_range[0])]/107770, width = width, facecolor=[0.9,0.9,0.9], edgecolor='k', alpha=1, label='ice')
        axs[ss].bar(bin_centers[(bin_centers<mix_range[1])&(bin_centers>mix_range[0])], 100*hist_out[0][(bin_centers<mix_range[1])&(bin_centers>mix_range[0])]/107770, width = width, facecolor=[255/255,165/255, 100/255], edgecolor='k', alpha=1, label='mix')
        axs[ss].bar(bin_centers[(bin_centers<sco_range[1])&(bin_centers>sco_range[0])], 100*hist_out[0][(bin_centers<sco_range[1])&(bin_centers>sco_range[0])]/107770, width = width, facecolor=[150/255,200/255,255/255], edgecolor='k', alpha=1, label='supercool')
        axs[ss].bar(bin_centers[(bin_centers<liq_range[1])&(bin_centers>liq_range[0])], 100*hist_out[0][(bin_centers<liq_range[1])&(bin_centers>liq_range[0])]/107770, width = width, facecolor=[40/255,140/255,255/255], edgecolor='k', alpha=1, label='liquid')
        if ss == 0:
            axs[ss].legend(fontsize=9)
    else:
        axs[ss].bar(bin_centers[(bin_centers<ice_range[1])&(bin_centers>ice_range[0])], 100*hist_out[0][(bin_centers<ice_range[1])&(bin_centers>ice_range[0])]/107770, width = width, facecolor='k', edgecolor='k', alpha=1, label='ice')
        axs[ss].bar(bin_centers[(bin_centers<mix_range[1])&(bin_centers>mix_range[0])], 100*hist_out[0][(bin_centers<mix_range[1])&(bin_centers>mix_range[0])]/107770, width = width, facecolor='k', edgecolor='k', alpha=1, label='mix')
        axs[ss].bar(bin_centers[(bin_centers<sco_range[1])&(bin_centers>sco_range[0])], 100*hist_out[0][(bin_centers<sco_range[1])&(bin_centers>sco_range[0])]/107770, width = width, facecolor='k', edgecolor='k', alpha=1, label='supercool')
        axs[ss].bar(bin_centers[(bin_centers<liq_range[1])&(bin_centers>liq_range[0])], 100*hist_out[0][(bin_centers<liq_range[1])&(bin_centers>liq_range[0])]/107770, width = width, facecolor='k', edgecolor='k', alpha=1, label='liquid')
    
    # plot labels
    axs[ss].set_title(f'{season}')
    axs[ss].set_xlim(-45,20)
    axs[ss].set_ylim(0,4)
    if ss == 0:
        axs[ss].set_ylabel('% of observations')
        axs[ss].set_xlabel('T ($^{\circ}$C)')
    if ss > 0:
        axs[ss].set_yticks([])
    


# save
if VERSION == 1:
    print(f'{ds.year[0].values}-{ds.year[-1].values}, {hours[0]}Z\nT when RH liq >= {rh_thresh_liq}')
    savename = f'./figures/T_RH{rh_thresh_liq}_seasonalrange_{ds.year[0].values}-{ds.year[-1].values}.png'
elif VERSION == 2:
    print(f'{ds.year[0].values}-{ds.year[-1].values}, {hours[0]}Z\nT when RH liq >= {rh_thresh_liq} and RH ice >= {rh_thresh_ice}')
    savename = f'./figures/T_RHice{rh_thresh_ice}_RHliq{rh_thresh_liq}_seasonalrange_{ds.year[0].values}-{ds.year[-1].values}.png'

print(f' >>> save as {savename}')
fig.savefig(savename, transparent=True, bbox_inches='tight', pad_inches=0, dpi=300)

```

### Calculate distribution of heights that different cloud types (temperature-defined) occur at in each season

```python
#====================================
# temperature ranges for cloud types
liq_range = (  0,  20)
sco_range = ( -5,  0 )
mix_range = (-25, -5 )
ice_range = (-60, -25)
#====================================

# create dictionary to store all
cloud_heigh_type = {}
for season in seasons:
    cloud_heigh_type[season] = {}
    
# run through all seasons
for ss, season in enumerate(seasons):

    liq_heights = np.array([])
    sco_heights = np.array([])
    mix_heights = np.array([])
    ice_heights = np.array([])
    
    # open seasonal data
    ds = xr.open_dataset(f'./data/{season}_1997-2016_v{VERSION}.nc')
    
    
    # find total number of days with observations
    # VERSION 1
    #----------
    if VERSION == 1:
        total_obs = (np.isnan(ds.RH)==False).sum(dim='doy').sum(dim='year').sum(dim='h').data
        
    # VERSION 2
    #----------
    elif VERSION == 2:
        # use RH_liq here since RH_ice values set to nan when T > 0
        total_obs = (np.isnan(ds.RH_liq)==False).sum(dim='doy').sum(dim='year').sum(dim='h').data
    
    cloud_heigh_type[season]['total_obs'] = total_obs
    
    
    # record number of times each cloud type is present in at least one place in a sounding
    num_liq = 0
    num_sco = 0
    num_mix = 0
    num_ice = 0
    num_sounding = 0
    
    # run through day by day and collect heights
    # where different cloud types might exist
    for year in ds.year:
        for doy in ds.doy:

            # grab heights with specific constraints
            ds_spec = ds.sel(year=year, doy=doy)

            
            # VERSION 1
            #----------
            if VERSION == 1:
                
                # liquid clouds
                liq_height = ds_spec.h.values[(ds_spec.T>=liq_range[0])&(ds_spec.T<liq_range[1])&(ds_spec.RH>=rh_thresh_liq)]
                # supercooled
                sco_height = ds_spec.h.values[(ds_spec.T>=sco_range[0])&(ds_spec.T<sco_range[1])&(ds_spec.RH>=rh_thresh_liq)]
                # mixed
                mix_height = ds_spec.h.values[(ds_spec.T>=mix_range[0])&(ds_spec.T<mix_range[1])&(ds_spec.RH>=rh_thresh_liq)]
                # ice only
                ice_height = ds_spec.h.values[(ds_spec.T>=ice_range[0])&(ds_spec.T<ice_range[1])&(ds_spec.RH>=rh_thresh_liq)]
            
            
            # VERSION 2
            #----------
    
            elif VERSION == 2:
                # liquid clouds, use liquid rh threshold and liquid RH value
                liq_height = ds_spec.h.values[(ds_spec.T>=liq_range[0])&(ds_spec.T<liq_range[1])&(ds_spec.RH_liq>=rh_thresh_liq)]
                # supercooled, use ice rh threshold and ice RH value
                sco_height = ds_spec.h.values[(ds_spec.T>=sco_range[0])&(ds_spec.T<sco_range[1])&(ds_spec.RH_ice>=rh_thresh_ice)]
                # mixed, use ice rh threshold and ice RH value
                mix_height = ds_spec.h.values[(ds_spec.T>=mix_range[0])&(ds_spec.T<mix_range[1])&(ds_spec.RH_ice>=rh_thresh_ice)]
                # ice only
                ice_height = ds_spec.h.values[(ds_spec.T>=ice_range[0])&(ds_spec.T<ice_range[1])&(ds_spec.RH_ice>=rh_thresh_ice)]
            
            
            # check that no heights are being double-assigned to different phases
            #--------------------------------------------------------------------
            if len(liq_height) > 0:
                for hh in liq_height:
                    if hh in sco_height or hh in mix_height or hh in ice_height:
                        print(f'ERROR: liquid cloud also classified as supercooled, mixed, or ice @ h={hh} on doy={doy}, {season} {year}')
                        break
            if len(sco_height) > 0:
                for hh in sco_height:
                    if hh in mix_height or hh in ice_height:
                        print(f'ERROR: supercooled cloud also classified as mixed or ice @ h={hh} on doy={doy}, {season} {year}')
                        break
            if len(mix_height) > 0:
                for hh in mix_height:
                    if hh in ice_height:
                        print(f'ERROR: mixed cloud also classified as ice @ h={hh} on doy={doy}, {season} {year}')
                        break
            
            # record whether cloud type is present in sounding
            if len(liq_height) > 0:
                num_liq += 1
            if len(sco_height) > 0:
                num_sco += 1
            if len(mix_height) > 0:
                num_mix += 1
            if len(ice_height) > 0:
                num_ice += 1
            
            # save to list
            liq_heights = np.append(liq_heights, liq_height)
            sco_heights = np.append(sco_heights, sco_height)
            mix_heights = np.append(mix_heights, mix_height)
            ice_heights = np.append(ice_heights, ice_height)
            
            # add 1 to num soundings unless profile was all nans
            
            # VERSION 1
            #----------
            if VERSION == 1:
                if np.sum(np.isnan(ds_spec.RH)) != len(ds_spec.RH):
                    num_sounding+=1
                    
            # VERSION 2
            #----------
            elif VERSION == 2:
                if np.sum(np.isnan(ds_spec.RH_liq)) != len(ds_spec.RH_liq):
                    num_sounding+=1
            
    # save each cloud type to dictionary
    cloud_heigh_type[season]['liquid'] = {}
    cloud_heigh_type[season]['supercool'] = {}
    cloud_heigh_type[season]['mix'] = {}
    cloud_heigh_type[season]['ice'] = {}
    
    # save distribution of heights
    cloud_heigh_type[season]['liquid']['heights'] = liq_heights
    cloud_heigh_type[season]['supercool']['heights'] = sco_heights
    cloud_heigh_type[season]['mix']['heights'] = mix_heights
    cloud_heigh_type[season]['ice']['heights'] = ice_heights
    
    # save number of soundings that contain each cloud type
    cloud_heigh_type[season]['liquid']['num'] = num_liq     # # soundings containing liquid cloud
    cloud_heigh_type[season]['supercool']['num'] = num_sco  # # soundings containing supercooled cloud
    cloud_heigh_type[season]['mix']['num'] = num_mix        # # soundings containing mixed cloud
    cloud_heigh_type[season]['ice']['num'] = num_ice        # # soundings containing ice cloud
    
    # save number of non-nan soundings
    cloud_heigh_type[season]['num_sound'] = num_sounding
    
```

###  Plot frequency of cloud occurence as a function of height
specify relative humidity threshold for clouds

```python

#==========================================================
seasons = ['winter', 'spring', 'summer', 'autumn']
colors = ['royalblue', 'dodgerblue', 'indianred', 'darkorange']
#==========================================================


low_level_freq = np.array([])
mid_level_freq = np.array([])
hig_level_freq = np.array([])

fig, ax = plt.subplots(figsize=(2,5))
assert len(hours) == 1, 'error in title with multiple hours'
ax.set_xlabel('frequency (%)')
ax.set_ylabel('height (km)')
ax.set_xlim(-2,77)
ax.set_xticks([0,25,50,75])

# low and mid - level clouds
ax.hlines(2, ax.get_xlim()[0], ax.get_xlim()[-1], colors='lightgray', alpha=0.5, zorder=-1)
ax.hlines(4, ax.get_xlim()[0], ax.get_xlim()[-1], colors='lightgray', alpha=0.5, zorder=-1)
ax.hlines(6, ax.get_xlim()[0], ax.get_xlim()[-1], colors='lightgray', alpha=0.5, zorder=-1)

for ss, season in enumerate(seasons):

    # open seasonal data
    ds = xr.open_dataset(f'./data/{season}_1997-2016_v{VERSION}.nc')
    
    
    # VERSION 1
    #----------
    if VERSION == 1:
        # find number of days when liquid RH above threshold
        above_thresh = (ds.RH >= rh_thresh_liq).sum(dim='doy').sum(dim='year')
    
        # find total number of days with observations
        # use RH_liq here since RH_ice set to nan when T > 0
        total_obs = (np.isnan(ds.RH)==False).sum(dim='doy').sum(dim='year')
    
    
    # VERSION 2
    #----------
    elif VERSION == 2:
        # find number of days when either liquid or ice RH is above their specified thresholds
        above_thresh = ((ds.RH_liq >= rh_thresh_liq)+(ds.RH_ice >= rh_thresh_ice)).sum(dim='doy').sum(dim='year')
    
        # find total number of days with observations
        # use RH_liq here since RH_ice set to nan when T > 0
        total_obs = (np.isnan(ds.RH_liq)==False).sum(dim='doy').sum(dim='year')
    
    # plot frequency of cloud occurence as a function of height
    freq = 100*above_thresh/total_obs
    if str(season) in ['spring', 'autumn']:
        linestyle = (0,(2,0.25))
        linewidth = 2
    else:
        linestyle = 'solid'
        linewidth = 2
    ax.plot(freq, ds.h, c=colors[ss], label=f'{season}', linestyle=linestyle, linewidth=linewidth)
    
    # calculate percentage of seasonal clouds found
    # in different height ranges
    low_freq = freq.isel(h=ds.h<2.05)
    mid_freq = freq.isel(h=(ds.h>2)&(ds.h<4.05))
    hig_freq = freq.isel(h=(ds.h>3.95))
    
    low_perc = (np.sum(low_freq)/np.sum(freq)).data
    mid_perc = (np.sum(mid_freq)/np.sum(freq)).data
    hig_perc = (np.sum(hig_freq)/np.sum(freq)).data

    low_level_freq = np.append(low_level_freq, low_perc)
    mid_level_freq = np.append(mid_level_freq, mid_perc)
    hig_level_freq = np.append(hig_level_freq, hig_perc)
    
    print(f'{season}: \n  - high-level (4-6 km): {hig_perc*100:.1f}%')
    print(f'  - mid-level (2-4 km): {mid_perc*100:.1f}%')
    print(f'  - low-level (0-2 km): {low_perc*100:.1f}%')
    
plt.legend(fontsize=10, loc=9, bbox_to_anchor=(1.4, 1))

if VERSION == 1:
    ax.set_title(f'{ds.year[0].values}-{ds.year[-1].values}, {hours[0]}Z\nRH liq >= {rh_thresh_liq}')
elif VERSION == 2:
    ax.set_title(f'{ds.year[0].values}-{ds.year[-1].values}, {hours[0]}Z\nRH liq >= {rh_thresh_liq}\nRH ice >= {rh_thresh_ice}')
    
# save
if VERSION == 1:
    savename = f'./figures/CloudFreq_RH{rh_thresh_liq}_{ds.year[0].values}-{ds.year[-1].values}_{hours}Z.png'
elif VERSION == 2:
    savename = f'./figures/CloudFreq_RHliq{rh_thresh_liq}_RHice{rh_thresh_ice}_{ds.year[0].values}-{ds.year[-1].values}_{hours}Z.png'
print(f' >>> save as {savename}')
fig.savefig(savename, transparent=True, bbox_inches='tight', pad_inches=0, dpi=300)



# plot fraction of clouds each season that occur 
# in different height ranges
fig, ax = plt.subplots(figsize=(4,4))
ax.bar(seasons, hig_level_freq, bottom=mid_level_freq+low_level_freq, color='k', label='4-6 km')
ax.bar(seasons, mid_level_freq, bottom=low_level_freq, color='gray', label='2-4 km')
ax.bar(seasons, low_level_freq, color='lightgray', label='0-2 km')

ax.set_ylabel('fraction of cloud heights')
fig.legend(loc=9, bbox_to_anchor=(1.075, 0.9))

# save
# savename = f'./figures/clouds_byheight_{ds.year[0].values}-{ds.year[-1].values}.png'
# print(f' >>> save as {savename}')
# fig.savefig(savename, transparent=True, bbox_inches='tight', pad_inches=0, dpi=300)
```

### Monthly cloud frequencies, depending on RH threshold

```python
#===================================================
# seasons = ['winter', 'autumn','winter', 'autumn']
seasons = ['winter', 'spring', 'summer', 'autumn']

rh_thresh_values = [110,105,100]
#===================================================

# define doy range for each month
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
month_range = {}
month_range['Jan'] = (0,32)
month_range['Feb'] = (32,60)
month_range['Mar'] = (60,91)
month_range['Apr'] = (91,121)
month_range['May'] = (121,152)
month_range['Jun'] = (152,182)
month_range['Jul'] = (182,213)
month_range['Aug'] = (213,244)
month_range['Sep'] = (244,274)
month_range['Oct'] = (274,305)
month_range['Nov'] = (305,335)
month_range['Dec'] = (335,366)

# dictionary to store monthly cloud frequencies
month_cloud_freq = {}
for rh_thresh in rh_thresh_values:
    month_cloud_freq[rh_thresh] = {}
    
# go through each rh value used to detect clouds
for rh_thresh in rh_thresh_values:
    
    
    month_freq = []
    
    # run through each seasonal data set
    for ss, season in enumerate(seasons):
        
        # open seasonal data
        ds = xr.open_dataset(f'./data/{season}_1997-2016_v{VERSION}.nc')
    
        # define month numbers
        month_num = 1+(ss*3)+np.array([0, 1, 2])

        # run through each month in season and find freq
        # of at least one cloud in each season
        for month in month_num:
            
            # crop seasonal data set to doy range associated with each month
            doy_range = month_range[months[month-1]]
            ds_month = ds.sel(doy=slice(doy_range[0],doy_range[1]))

            
            # VERSION 1
            #----------
            if VERSION == 1:
                # find number of days where at least one cloud is present in each sounding
                above_thresh_persounding = ((ds_month.RH >= rh_thresh).sum(dim='h')>0).sum(dim='doy').sum(dim='year')

                # find total number of non-nan soundings
                # # nans varies slightly with height (by 1 or 2 out of 1700+), os just
                # ignore h-dependence and grab total # obs as # non-nan obs @ lowest level
                total_obs = (np.isnan(ds_month.RH)==False).sum(dim='doy').sum(dim='year')[0]
        
            # VERSION 2
            #----------
            elif VERSION == 2:
                
                # specify ice or liquid!
                #=======================
                VAR = 'ice'
#                 VAR = 'liq'
#                 VAR = 'both'
                #=======================
    
                # find number of days where at least one cloud is present in each sounding
                if str(VAR) == 'ice':
                    above_thresh_persounding = ((ds_month.RH_ice >= rh_thresh).sum(dim='h')>0).sum(dim='doy').sum(dim='year')
                elif str(VAR) == 'liq':
                    above_thresh_persounding = ((ds_month.RH_liq >= rh_thresh).sum(dim='h')>0).sum(dim='doy').sum(dim='year')
                elif str(VAR) == 'both':
                    above_thresh_persounding = (((ds_month.RH_liq >= rh_thresh)+(ds_month.RH_ice >= rh_thresh)).sum(dim='h')>0).sum(dim='doy').sum(dim='year')
                    
                # find total number of non-nan soundings
                # # nans varies slightly with height (by 1 or 2 out of 1700+), os just
                # ignore h-dependence and grab total # obs as # non-nan obs @ lowest level
                total_obs = (np.isnan(ds_month.RH_liq)==False).sum(dim='doy').sum(dim='year')[0]


            # plot frequency of 1+ cloud occurence(s) somewhere in 0-6 km
            freq = above_thresh_persounding/total_obs

            month_freq.append(freq.data)
        
    # save to dictionary
    month_cloud_freq[rh_thresh] = month_freq
    

if VERSION == 2:
    print(f'for {VAR} cloud types')
    
    
```

#### Plot

```python

fig, ax = plt.subplots(figsize=(5.85,5))

# run through all rh threshold values and plot monthly cloud frequency
for rr, rh_thresh in enumerate(rh_thresh_values):
    
    if VERSION == 1:
        LABEL = f'{rh_thresh_liq}'
        
    elif VERSION == 2:
        LABEL = f'RHliq = {rh_thresh_liq}, RHice = {rh_thresh_ice}'
    
    ax.plot(months, month_cloud_freq[rh_thresh], lw=3, label=LABEL, marker='o', c=[238/255,99/255,99/255], clip_on=False)
    
    # label each curve
    ax.text(months[-1], month_cloud_freq[rh_thresh][-1], f'     RH = {rh_thresh}', 
            c=[238/255,99/255,99/255], verticalalignment='center', )
# ax.set_xlim('Jan', 'Dec')
ax.set_ylim(0,1)

# save
savename = f'./figures/AnyCloud_occurence_{ds.year[0].values}-{ds.year[-1].values}.png'
print(f' >>> save as {savename}')
fig.savefig(savename, transparent=True, bbox_inches='tight', pad_inches=0, dpi=300)
```

### Seasonal cloud frequencies 2006-2011, with seasons re-defined to match [Liu et al., (2012)](https://digitalcommons.unl.edu/usdeptcommercepub/344/)

winter: Dec, Jan, Feb
spring: Mar, Apr, May
summer: Jun, Jul, Aug
autumn: Sep, Oct, Nov

using a jet-like colormap :/ to match the colormap in the publication

```python

#===================================================
seasons = ['winter', 'spring', 'summer', 'autumn']
# seasons = ['winter', 'autumn','winter', 'autumn']
rh_thresh_values = [85, 90, 95]
#===================================================

# define doy range for each month
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
month_range = {}
month_range['Jan'] = (0,32)
month_range['Feb'] = (32,60)
month_range['Mar'] = (60,91)
month_range['Apr'] = (91,121)
month_range['May'] = (121,152)
month_range['Jun'] = (152,182)
month_range['Jul'] = (182,213)
month_range['Aug'] = (213,244)
month_range['Sep'] = (244,274)
month_range['Oct'] = (274,305)
month_range['Nov'] = (305,335)
month_range['Dec'] = (335,366)

# dictionary to store monthly cloud frequencies
month_cloud_freq = {}
for rh_thresh in rh_thresh_values:
    month_cloud_freq[rh_thresh] = {}
    
# go through each rh value used to detect clouds
for rh_thresh in rh_thresh_values:
    
    
    month_freq = []
    
    # run through each seasonal data set
    for ss, season in enumerate(seasons):
        
        # open seasonal data
        ds = xr.open_dataset(f'./data/{season}_1997-2016_v{VERSION}.nc')

        # define month numbers
        month_num = 1+(ss*3)+np.array([0, 1, 2])

        # run through each month in season and find freq
        # of at least one cloud in each season
        for month in month_num:
            
            # crop seasonal data set to doy range associated with each month
            doy_range = month_range[months[month-1]]
            ds_month = ds.sel(doy=slice(doy_range[0],doy_range[1]))#.sel(year=slice(2006,2011))

            # VERSION 1
            #----------
            if VERSION == 1:
                # find number of days where at least one cloud is present in each sounding
                above_thresh_persounding = ((ds_month.RH >= rh_thresh).sum(dim='h')>0).sum(dim='doy').sum(dim='year')

                # find total number of non-nan soundings
                # # nans varies slightly with height (by 1 or 2 out of 1700+), os just
                # ignore h-dependence and grab total # obs as # non-nan obs @ lowest level
                total_obs = (np.isnan(ds_month.RH)==False).sum(dim='doy').sum(dim='year')[0]

            
            # VERSION 2
            #----------
            elif VERSION == 2:
                # find number of days where at least one cloud is present in each sounding
                above_thresh_persounding = ((ds_month.RH_liq >= rh_thresh).sum(dim='h')>0).sum(dim='doy').sum(dim='year')
                
                # find total number of non-nan soundings
                # # nans varies slightly with height (by 1 or 2 out of 1700+), os just
                # ignore h-dependence and grab total # obs as # non-nan obs @ lowest level
                total_obs = (np.isnan(ds_month.RH_liq)==False).sum(dim='doy').sum(dim='year')[0]

            # plot frequency of 1+ cloud occurence(s) somewhere in 0-6 km
            freq = above_thresh_persounding/total_obs

            month_freq.append(freq.data)
        
    # save to dictionary
    month_cloud_freq[rh_thresh] = month_freq


    
# find seasonal averages
RH_THRESH = 90
WINT_MEAN = 100*np.nanmean(np.append(month_cloud_freq[RH_THRESH][:2], month_cloud_freq[RH_THRESH][-1]))
SPRI_MEAN = 100*np.nanmean(month_cloud_freq[RH_THRESH][2:5])
SUMM_MEAN = 100*np.nanmean(month_cloud_freq[RH_THRESH][5:8])
AUTU_MEAN = 100*np.nanmean(month_cloud_freq[RH_THRESH][8:11])

# plot
fig, ax = plt.subplots(figsize=(5,5))
colors = ax.scatter(['DJF', 'MAM', 'JJA', 'SON'], [WINT_MEAN, SPRI_MEAN, SUMM_MEAN, AUTU_MEAN], 
            c=[WINT_MEAN, SPRI_MEAN, SUMM_MEAN, AUTU_MEAN], s=100, edgecolor='k',
            cmap='gist_rainbow_r', vmin=40, vmax=105)
plt.colorbar(colors)
ax.set_ylabel('cloud frequency (%)')
ax.set_xlabel('season')

# save
savename = f'./figures/shifted_months_2006_2011_cloudfreq.png'
print(f' >>> save as {savename}')
fig.savefig(savename, transparent=True, bbox_inches='tight', pad_inches=0, dpi=300)
```
