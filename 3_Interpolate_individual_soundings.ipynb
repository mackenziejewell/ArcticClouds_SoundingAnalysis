### Calculating relative humidity ($RH$, %) from $T$, $T_{d}$, $p$

***
Originally, I used [relative_humidity_from_dewpoint(temperature, dewpoint)](https://unidata.github.io/MetPy/latest/api/generated/metpy.calc.relative_humidity_from_dewpoint.html) to calculate the relative humidity. This function uses temperature $(\mathrm{T})$ and dewpoint $(\mathrm{T_d})$ to calculate relative humidity $(\mathrm{RH})$ as the ratio of vapor pressure $(\mathrm{e})$ to saturation vapor pressure $(\mathrm{e_s})$.

$$RH = 100 \; \times \; \frac{e(T_{d})}{e_{s}(T)}$$

This is an approximation as in actuality $RH \approx 100 \; \times \; \frac{e(T_{d})}{e_{s}(T)}$. And further, this only specifies the realtive humidity with respect to liquid water, not accounting for whether the atmoosphere is saturated with respect to ice. This is important here, because usually $\mathrm{T < 0 \; ^{\circ}C}$ in these Arctic soundings, and the equilibrium vapor pressure for ice is lower than that for liquid water droplets. So, if liquid water dorplets are forming given the pressure and temperature, and there are sufficient ice nuclei, ice will also be forming. 

***
So below, I use some of my own functions to solve for $(\mathrm{RH})$ with respect to both liquid and ice using equations from the literature (primarily Rogers and Yau, 1996). The specific equations from this text are labeled with R&Y and the equation number. We want to be able to calculate $\mathrm{RH}$ from only variables provided in the sounding data, which include $T$, $T_{d}$, $p$ (pressure).

$\mathrm{RH}$ can be calculated from mixing ratio $(\mathrm{w})$ and saturation mixing ratio $(\mathrm{w_s})$ as:

$$RH = 100 \; \times \frac{w}{w_{s}} \tag{R&Y 2.20}$$


We can use the metpy.calc package to calculate $w(q)$ from specific humidity $q$ using [mixing_ratio_from_specific_humidity(specific_humidity)](https://unidata.github.io/MetPy/latest/api/generated/metpy.calc.mixing_ratio_from_specific_humidity.html). $q(p, T_{d})$ can be calculated from $p$ and $T_{d}$ using [specific_humidity_from_dewpoint(pressure, dewpoint)](https://unidata.github.io/MetPy/latest/api/generated/metpy.calc.specific_humidity_from_dewpoint.html).
    
    
$w_{s}$ can be calculated as:

$$w_s = \epsilon \frac{{e}_{s}(T)}{p-{e}_{s}(T)} \tag{R&Y 2.18}$$

where $\epsilon \approx 0.622$ and ${e}_{s}(T)$ is the saturation vapor pressure either for ice or water. I have written a function saturation_vaporpressures_from_temperature in LIB_sounding_analysis to calculate this for water and for ice using different formulations from literature. Alduchov and Eskridge (1996) or Huang (2018). See LIB_sounding_analysis or run help(saturation_vaporpressures_from_temperature) for more information.




***

Alduchov, O. A., and R. E. Eskridge, 1996: Improved Magnus Form Approximation of Saturation Vapor Pressure.  J. Appl. Meteor. Climatol., 35, 601–609, https://doi.org/10.1175/1520-0450(1996)035<0601:IMFAOS>2.0.CO;2
    
Huang, J., 2018: A Simple Accurate Formula for Calculating Saturation Vapor Pressure of Water and Ice. J. Appl. Meteor. Climatol., 57, 1265–1272, https://doi.org/10.1175/JAMC-D-17-0334.1.

Yau, M. K., and R. R. Rogers. A Short Course in Cloud Physics, Elsevier Science & Technology, 1996. ProQuest Ebook Central, https://ebookcentral.proquest.com/lib/osu/detail.action?docID=4011796, EBOOK ISBN: 9780080570945

### Interpolate and plot data

```python
#===============================
max_height = 10*units.kilometer
min_height =  0*units.kilometer
bin_width = 0.1*units.kilometer
#===============================

[new_RH_liq, new_RH_ice, new_T], new_H, edges = interpolate_soundings(variable = [RH_liq, RH_ice, T], heights = H, 
                                                      bin_width = bin_width, min_height = min_height, 
                                                      max_height = max_height, method = 'mean', 
                                                      suppress_plots = False, suppress_prints = True)

# which initial and interpolated
# RH data to plot
#===============================
# initial_RH = RH_liq
# interpd_RH = new_RH_liq

initial_RH = RH_ice
interpd_RH = new_RH_ice
#===============================

fig, ax = plt.subplots(figsize=(3, 8))
# ax.plot(RH, H, c='k', zorder=0)
temp = ax.scatter(initial_RH[H<=max_height], H[H<=max_height], c=T[H<=max_height], edgecolor='k', s=100, vmin=-40, vmax = 0, zorder=1, label='sounding data')
plt.colorbar(temp)

RH_thresh = 100
ax.vlines(RH_thresh,-1,20, colors='k', alpha=0.5, zorder=-1)

plt.xlim(-5,round(np.max(interpd_RH).magnitude)+5)
plt.ylim(-0.25,round(np.max(new_H).magnitude)+0.25)


for edge in edges:
    ax.hlines(edge,-5, 140, colors='k', alpha=0.1, zorder=-1, label='bin edges')

ax.plot(interpd_RH, new_H, c='r', zorder=1)
ax.scatter(interpd_RH, new_H, c=new_T, edgecolor='r', marker='D', s=20, vmin=-40, vmax = 0, zorder=1, label='interpolated data')

handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(), bbox_to_anchor=(1,1.2))
```
